```python
from pandas._libs.tslibs import conversion
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
import numpy as np

url = '1.01. Simple linear regression.csv'
data = pd.read_csv(url)
x1 = data['SAT']
y = data['GPA']

plt.scatter(x1, y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()

data.hist()
plt.show()
conversion = data.corr()
```

```python
regression_model = linear_model.LinearRegression()
regression_model.fit(x1,y)
print(regression_model.intercept_)
print(regression_model.coef_)
```

```python
url = 'mtcars.csv'
data = pd.read_csv(url)
data.hist()
conversion = data.corr()
print(conversion)
```

```python
X = data.drop(data.columns[[0,2]], axis = 1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'blueviolet')
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']), y = data['mpg']) #dạy nó học
print(model.intercept_)
print(model.coef_)

```

```python
model.score(X=pd.DataFrame(data['wt']), y = data['mpg']) #do chinh xac
```

```python
y_pred = model.predict(X=pd.DataFrame(data['wt'])) 
residual = data['mpg'] - y_pred #sai so
residual.describe()
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'red')
plt.plot(data['wt'], y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2]).T
y = data['mpg']
model.fit(X, y)
print('Intercept của mô hình bậc 2: ')
print(model.intercept_)
print('coeffient của mô hình bậc 2: ')
print(model.coef_)
print('Độ chính xác của mô hình: ')
print(model.score(X,y))
```

```python
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor) 
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xlim = (0,7))
plt.plot(range, y_pred, color = 'red')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3]).T # a1x + a2x^2 +b, T nghịch đảo
y = data['mpg']
model.fit(X, y)
print('Intercept của mô hình bậc 3: ')
print(model.intercept_)
print('coeffient của mô hình bậc 3: ')
print(model.coef_)
print('Độ chính xác của mô hình: ')
print(model.score(X,y))
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict(X = predictor) 
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xlim = (0,7))
plt.plot(range, y_pred, color = 'green')
```

```python
# vẽ đồ thị bậc 3 đến bậc 10
#xoa cot 0 den 2
X = data.drop(data.columns[[0, 2]], axis=1)
y = data['mpg'] 

arr = np.arange(3, 11)
range1 = np.arange(1.5,5.5,0.1)
for id in arr:

    a_range = np.arange(1, i + 1)
    
    X_1 = pd.DataFrame([X['wt'] ** i for i in a_range]).T
    model = linear_model.LinearRegression()
    model.fit(X_1, y)

    print(f'Intercept của mô hình bậc {i}: {model.intercept_}')
    print(f'Coefficients của mô hình bậc {i}: {model.coef_}')
    print(f'Độ chính xác của mô hình bậc {i}: {model.score(X,y)}')

    
    predictor = pd.DataFrame([range,range**2, range**i]).T
    y_pred = model.predict(X = predictor)

    data.plot(kind = 'scatter', x = 'wt',y = 'mpg',figsize = (9,9),color = 'black',xlim  = (0,9))
    plt.plot(range,y_pred,color = 'blue')
    plt.title(f'Mô hình đa thức bậc {i}')
    plt.show()
```
